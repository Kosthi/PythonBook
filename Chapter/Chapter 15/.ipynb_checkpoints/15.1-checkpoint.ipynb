```python
import matplotlib.pyplot as plt
```

```python
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

plt.style.use('fast')
fig, ax = plt.subplots()
ax.plot(input_values, squares, linewidth=3)

ax.set_title("X", fontsize=24)
ax.set_xlabel("Values", fontsize=24)
ax.set_ylabel("Y", fontsize=24)

ax.tick_params(axis='both', labelsize=14)
plt.show()
```

```python
x_values = [i for i in range(1, 1001)]
y_values = [x**2 for x in x_values]
plt.style.use('fast')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s=10, c=y_values, cmap=plt.cm.Reds)
# ax.plot(input_values, squares, linewidth=3)

ax.set_title("X", fontsize=18)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Y", fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.axis([0, 1100, 0, 1100000])
#plt.show()
plt.savefig('a.png', bbox_inches='tight')
```

```python
x_values = [i for i in range(1, 5001)]
y_values = [x**3 for x in x_values]
plt.style.use('fast')
fig, ax = plt.subplots(figsize=(15, 9))
ax.scatter(x_values, y_values, s=5, c=y_values, cmap=plt.cm.rainbow).set_antialiased(True)
# ax.plot(input_values, squares, linewidth=3)
#plt.set_antialiased(True) 
ax.set_title("X", fontsize=18)
ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Y", fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.axis([0, 5100, 0, 130000000000])
#plt.show()
plt.savefig('b.png', bbox_inches='tight')
```
